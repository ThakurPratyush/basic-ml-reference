```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# preparing data
X = 6 * np.random.rand(100,1) - 3  # (100,1) is shape of output vector and -3 for outliers
y = 0.5 * X**2 + 1.5 * X +2 + np.random.randn(100,1)
```

```python
plt.scatter(X,y,color = 'g')
plt.xlabel("x dataset")
plt.ylabel("y dataset")
```

```python
# non linear realtionship in between x and y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25, random_state= 42)
```

```python
## lets implement simple lr
from sklearn.linear_model import LinearRegression
```

```python
regression_1 = LinearRegression()
regression_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,regression_1.predict(X_test))
```

```python
print(score)
```

```python
# plotting the data 
plt.scatter(X_train,y_train)
plt.plot(X_train,regression_1.predict(X_train), color = 'r')
plt.xlabel("X")
plt.ylabel("y")
```

```python
from sklearn.pipeline import Pipeline
```

```python
## this function would take dregree as input and based on that would draw line on data
```

```python
def polynomial_regression(degree):
    X_new = np.linspace(-3,3,200).reshape(200,1)
    poly_features = PolynomialFeatures(degree = degree ,include_bias = True)
    lin_reg = LinearRegression()
    pol_reg_pipeline = Pipeline([
       ("poly_regression", poly_features),
       ("linear_reg",lin_reg)
    ])
    pol_reg_pipeline.fit(X_train,y_train) ## firstly would get te polynomial features and then fit into the model
    y_pred_new = pol_reg_pipeline.predict(X_new)
    plt.plot(X_new, y_pred_new,'r', label = "degree "+str(degree),linewidth= 2 )
    plt.scatter(X_train, y_train,color = 'b')
    plt.scatter(X_test, y_test,color ='g')
    plt.legend(loc = 'upper left')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
polynomial_regression(10)
```
