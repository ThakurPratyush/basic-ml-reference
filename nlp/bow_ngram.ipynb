```python
import pandas as pd
import numpy as np
messages = pd.read_csv("smsspamcollection/SMSSpamCollection",sep = "\t", names = ["label","messages"])
```

```python
## datacleaning and preprocessing
```

```python
import re
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]' ,' ', messages['messages'][i])
    review = review.lower()
    review = review.split()   # list made of each review
    review = [ps.stem(word) for word in review  if  word not  in stopwords.words('english')]
    ## creating list of words for every review
    review = " ".join(review)
    corpus.append(review)
    
    
```

```python
#countvectorizer used for bow model
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 100)

#max_featuers : based on frequency we can take top 2500 features, some cases very rare occuring words dont play imp role
## take top 2500 words of max freq
#lowercase and stopwrods can be managed from here also
```

```python
X = cv.fit_transform(corpus).toarray()
```
