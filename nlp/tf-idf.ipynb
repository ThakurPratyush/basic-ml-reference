```python
import pandas as pd
import numpy as np
messages = pd.read_csv("smsspamcollection/SMSSpamCollection",sep = "\t", names = ["label","messages"])
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
import re
```

```python
import nltk
nltk.download("stopwords")
```

```python
 # nltk.download('wordnet')
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['messages'][i])
    review = review.lower()
    review = review.split()
    review = [wl.lemmatize(word) for word  in review if word not in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features = 100)
X =tfidf.fit_transform(corpus).toarray()
```
