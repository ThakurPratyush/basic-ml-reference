```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples= 750 , factor = 0.3, noise = 0.1)
## factor controls the separation in between the circles 
## noise is errous points 
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], c = y)
## we cannot use y obviously for a clusterin problem
```

```python
dbscan = DBSCAN(eps = 0.10)
dbscan.fit(X)
print(dbscan.fit_predict(X))
print(dbscan.labels_)
```

```python
# -1 outliers
```

```python
plt.scatter(X[:,0], X[:,1],c = dbscan.labels_)
```

```python
### a large number of clusters
```

```python
## corr to -1 are the outliers
import numpy as np
```

```python
outliers = np.where(dbscan.labels_ <0)
```

```python
outliers
```

```python
plt.scatter(X[:,0], X[:,1],c = dbscan.labels_)

#plt.scatter(X[outliers,0], X[outliers,1], c= "black")

plt.scatter(X[outliers,0], X[outliers,1], edgecolors = 'orange')
```
