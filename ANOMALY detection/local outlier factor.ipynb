```python
## Local outlier factor
```

points away from all clusters are global outliter but away from a particular cluster are outliers to that particular 
cluster called: local outliers

take a point with a k , say  k = 5 
then find avg distance of this points to k nearest neighbours 
then further find avg distances of these 5 points to the 5 neighbours and calculate the avg

if the higher avg distance is corr to point chosen earlier then can be considered outlier(loca)

uses knn inside it

```python
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples= 750 , factor = 0.3, noise = 0.1)
```

```python
plt.scatter(X[:,0], X[:,1], c = y)
```

```python
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = clf.fit_predict(X)
```

```python
y_pred
#-1 give outliers
X_scores = clf.negative_outlier_factor_
```

```python
import numpy as np
outliers = np.where(y_pred <0)
```

```python
# X_scores
plt.scatter(X[:,0], X[:,1],c = y_pred)

#plt.scatter(X[outliers,0], X[outliers,1], c= "black")

plt.scatter(X[outliers,0], X[outliers,1], edgecolors = 'orange')
```

```python
negative_outlier_factors = clf.negative_outlier_factor_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=negative_outlier_factors, cmap='coolwarm', edgecolor='k')
plt.colorbar(label='Negative Outlier Factor')
plt.title('Local Outlier Factor Scores')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
## coolwarm gives us the tendecy where cooler is the one which is more outliers
```
