### hyperparam tuning and cv

```python
## GRID SEARCH CV
```

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100,10,1,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params = dict(penalty= penalty , C = c_values, solver = solver)
## while creating the dict use the correct terms for k=keys
```

```python
## these are the keys value pairs in grid search all combinations of these would be tested one by one
## and best one selected
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator= model, param_grid = params , scoring = 'accuracy',cv = cv,n_jobs = -1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
## test each and all params
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
cr = classification_report(y_pred, y_test)
print(cr)
```

## logistic regression for imbalanced dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 1000, n_features =2 ,n_clusters_per_class =1,weights = [0.99],
                          random_state = 10,n_informative = 2,n_redundant = 0)


# cluster per class 1 suggests that for each cluster data would be coming from a single cluster only
                          
```

```python
Counter(y)
## 0 has 99 percent mroe data
```

```python
import seaborn as sns
```

```python
## we would print the both features and see where do these get assigned
sns.scatterplot(x = pd.DataFrame(X)[0],y = pd.DataFrame(X)[1], hue = y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y, test_size = 0.25,random_state =42)
```

```python
## hyper param tuning 
```

```python
## one of main attribute in case of imbalanced classes is classweights
# balanced means it gives equal wight to both th eclasses or we can specify in dictionary the weight we want to give to clases
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [0.5, 1, 10 ,50, 0.05, 10, 100]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
class_weight = [{0:w, 1:v} for w in [1, 10,50, 100] for v in [1,10,50,100]]
## since the attribte is type dictionary where we need to makelist of distionariers
params = dict(penalty = penalty , C = c_values, solver = solver, class_weight = class_weight)
```

```python
class_weight
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator = model, param_grid = params,scoring = 'accuracy',cv =cv)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
cr = classification_report(y_pred, y_test)
print(cr)
```

## logistic regression ROC

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot
```

```python
X,y = make_classification(n_samples = 1000, random_state = 1 , n_classes =2)
## split into train_TEst split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42)
```

```python
## dummy model that always predict 0 irrespective of input
dummy_model_prob = [0 for x in range(len(y_test))]
dummy_model_prob[:10]
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model_prob = model.predict_proba(X_test)
```

```python
model_prob
```

```python
## Lets focus on one output , here we are chosing the ouput of prob corresponsing to class 1
model_prob = model_prob[:,1]
```

```python
# model_prob
```

```python
## lets calcualte the scores
dummy_model_acc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)
print(dummy_model_acc)
print(model_auc)
```


###############################################


false posstive rate = FP/N = FP/TN+FP

TRUE POSTIVE RATE =
TP/P = TP/FN+TP


TPR IN Y AXIS and FPR in x axis we get roc curve

```python
dummy_fpr,dummy_tpr, _= roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr,th = roc_curve(y_test, model_prob)
```

```python
model_fpr.size,th.size
```

```python
pyplot.plot(dummy_fpr, dummy_tpr, linestyle = '--', label = 'Dummy Model')
pyplot.plot(model_fpr, model_tpr, marker = '.', label = 'logistic')
## axis labels 
pyplot.xlabel('false positive rate')
pyplot.ylabel('true postiive rate')
## show legend
pyplot.legend()
## show the plot
pyplot.show()
```

```python
## logitic regression works better than the DEFAULT MODEL SINCE more area under cuver
## while calculating the curvr we used thresholds, these are teh threshold based on which we decide the class for ex if 0.5 then > 0,5 called the
## postive class
```

```python
## we need to choose threshold in a way so that the true positibve rate is high and falso positive rate is low 
```

```python
## for each threshold we keep on plotting the curve and we choose that thresold for which we have the high tpr and low fpr
```

```python
import numpy as np
fig = pyplot.figure(figsize = (20,50))
ax = fig.add_subplot(111)
pyplot.plot(dummy_fpr, dummy_tpr, linestyle = '--', label = 'Dummy model')
pyplot.plot(model_fpr, model_tpr, marker = '.', label = 'Logistic')
for xyz in zip(model_fpr, model_tpr,th):
    ax.annotate('%s' % np.round(xyz[2],2), xy = (xyz[0], xyz[1]))
pyplot.xlabel('false positive rate')
pyplot.ylabel('true neg rate')
pyplot.show()
```
