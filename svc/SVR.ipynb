```python
## label and OHE 
## use label enocding where two categories only 
## for more than 2 we use label encoding
df.columns
```

```python
## inde and dep features
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train, y_test = train_test_split(X, y , test_size = 0.25, random_state = 42)
```

```python
X_test.head()
```

```python
## do model division before perfomring the encoding to avoid data leakage
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_train['smoker'] = le1.fit_transform(X_train['smoker'])
X_train['sex'] = le2.fit_transform(X_train['sex'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['smoker'] = le1.transform(X_test['smoker'])
X_test['sex'] = le2.transform(X_test['sex'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers = [('onehot' ,OneHotEncoder(drop = 'first') ,[3] )],remainder = 'passthrough')
## drop is first removes the column which was there before encoding
## reaminder = passthough would keep rest of columns while drop would remove rest of columns
```

```python
import sys
import numpy 
numpy.set_printoptions(threshold = sys.maxsize)
## to see the matrix created complelety
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
