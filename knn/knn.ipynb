```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(
    n_samples = 1000,
    n_features = 3,
    n_redundant = 1,
    n_classes = 2,
    random_state = 99
)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, algorithm = 'auto')
# p= 1 manhatan p =2 eucledian (default) , algorithm = 'kdtree', ball, auto (decides best automatically)
# wights uniform (all points given equal weight  , distance : inverse of distacnce, udf
# 
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
model = KNeighborsClassifier(algorithm='auto')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 'n_neighbors':[3,5,2,1,10,13],
                'weights' : ['uniform','distance'],
                'p' :  [1,2]
             }

```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator = model,param_grid = param_grid, refit = True, cv = cv)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```
